```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import joblib
from sklift.models import SoloModel
from sklift.viz import plot_qini_curve
from sklearn.ensemble import RandomForestClassifier

# import preprocessed train dataset
df_filtered = pd.read_csv('train_dataset.csv')
df_filtered = df_filtered.sort_index(axis=1)

y = df_filtered['activation']
treat = df_filtered['ind_recommended']
X = df_filtered.drop(['merchant','customer','activation',
                      'ind_recommended'], axis=1)

# RandomForestClassifier as the baseline estimator
rf_clf = RandomForestClassifier(
    n_estimators=300,
    max_depth=20,
    random_state=42,
    n_jobs=-1
)

# Initialize the SoloModel with the RandomForestClassifier
slearner_rf = SoloModel(estimator=rf_clf)

# Train the RandomForest slearner
slearner_rf.fit(X, y, treatment=treat)
```
